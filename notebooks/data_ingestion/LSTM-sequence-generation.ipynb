```python
data_subset = data[data['cell_id']=='018cf1a1-f945-4097-9c47-0c4690538bb5'].copy()
```

```python
data_subset['date'] = pd.to_datetime(data_subset['date'])
```

```python
#Get years in dataset
years = []
for i in data_subset['date']:
    if i.year not in years:
        years.append(i.year)
```

```python
years
```

```python
#If we want to drop columns from sequence
drop_columns = []
drop_indexes = []
for i in columns:
    if i in drop_columns:
        drop_indexes.append(columns.index(i))
```

```python
sequences = pd.DataFrame(columns=['cell_id','start_date','end_date','sequence'])
seq_length = 5
for year in years:
    #Get samples from cell for each year
    year_data = data_subset[data_subset['date'].dt.year == year].copy()
    #Reset Index
    year_data=year_data.reset_index(drop=True)
    
    #End Index
    end = year_data.last_valid_index() - seq_length
    
    #Go down the index generating sequences
    for i in range(0,end+1):
        sequence = []
        #Generate sequences based on seq_length, last index being "Goal"
        for j in range(0,seq_length+1):
            #Get row as list
            seq_item = year_data.iloc[i+j].values.flatten().tolist()
            
            print(seq_item[12])
            
            #Code for dropping items in row
            #Disabled for now
            #for inx in drop_indexes:
            #    del seq_item[index]
            
            sequence.append(seq_item)
        
        #Get other row info
        cell = sequence[0][columns.index('cell_id')]
        start_date = sequence[0][columns.index('date')]
        end_date = sequence[-1][columns.index('date')]
        
        row = pd.DataFrame(data = [[cell,start_date,end_date,sequence]], columns=['cell_id','start_date','end_date','sequence'])
        
        sequences = pd.concat([sequences,row],axis=0)

            
sequences = sequences.reset_index(drop=True)
```
